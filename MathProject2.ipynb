```python
def fixedPoint(func, a, b, tolerance):

    g_x = func
    xn_1 = a
    xn = g_x(xn_1)
    while abs(xn - xn_1) > tolerance:
      xn_1 = xn
      xn = g_x(xn_1)
    return xn
    raise NotImplementedError()
```

```python
assert np.isclose(fixedPoint(eq1, 2, 3, 1e-5), 2.1937309384071364)
```
